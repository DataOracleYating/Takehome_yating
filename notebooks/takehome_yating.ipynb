# 3. data visualization

## 3.1 Percentage of samples resistant (R) and sensitive (S) to each antibiotic

```python

# Creating empty lists to store the counts
antibiotics = []
resistant_percent = []
sensitive_percent = []

# Iterating over each antibiotic
for i in range(1, 18):
    antibiotic_column = f"Antibiotic_{i}"
    count = df[antibiotic_column].value_counts(normalize=True)  # Getting the percentage
    antibiotics.append(antibiotic_column)
    resistant_percent.append(count['R']*100)
    sensitive_percent.append(count['S']*100)

# Plotting the results
antibiotic_df = pd.DataFrame({'Antibiotics': antibiotics, 'Resistant': resistant_percent, 'Sensitive': sensitive_percent})

antibiotic_df
```

```python
# This code extracts the numerical part from the 'Antibiotics' column, sorts the DataFrame
# by the extracted numerical values, and plots a horizontal stacked bar chart showing
# the percentage of samples resistant (R) and sensitive (S) to each antibiotic.



# Extract the numerical part from 'Antibiotics' column and convert it to int
antibiotic_df['Antibiotics_num'] = antibiotic_df['Antibiotics'].str.extract('(\d+)').astype(int)

# Sort DataFrame by 'Antibiotics_num' in descending order
antibiotic_df = antibiotic_df.sort_values('Antibiotics_num', ascending=False)

# Remove the temporary 'Antibiotics_num' column
antibiotic_df.drop('Antibiotics_num', axis=1, inplace=True)

# Plotting the results
# Defining the colors for the plots
palette = ['#98CAF2', '#EC8DA3','#3A4EA7', '#BB5B6A']
antibiotic_df.plot(x="Antibiotics", 
                   y=["Resistant", "Sensitive"], 
                   kind="barh", 
                   stacked=True, 
                   color=palette[:2]) 
# Using the colors '#98CAF2' and '#EC8DA3' for 'Resistant' and 'Sensitive' respectively in the plot.

plt.xlabel('Percentage')
plt.title('Percentage of samples resistant (R) and sensitive (S) to each antibiotic')

# Moving the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show();

# Using a horizontal stacked bar chart to compare the percentage of resistant and sensitive
# samples for each antibiotic. Resistant and sensitive percentages are stacked for easy comparison.
```

## 3.2 Presence of each resistance gene across the samples

```python
# Creating empty lists to store the counts
genes = []
present = []
absent = []

# Iterating over each gene
for i in range(1, 6):
    gene_column = f"Gene_{i}"
    count = df[gene_column].value_counts()  # Getting the count
    genes.append(gene_column)
    present.append(count[1])
    absent.append(count[0])

# Creating a DataFrame from the lists
gene_df = pd.DataFrame({"Genes": genes, "Present": present, "Absent": absent})

# Extract the numerical part from 'Genes' column and convert it to int
gene_df['Gene_num'] = gene_df['Genes'].str.extract('(\d+)').astype(int)

# Sort DataFrame by 'Gene_num' in descending order
gene_df = gene_df.sort_values('Gene_num', ascending=False)

# Remove the temporary 'Gene_num' column
gene_df.drop('Gene_num', axis=1, inplace=True)

gene_df

```

```python
# Plotting the results
palette = ['#98CAF2', '#EC8DA3','#3A4EA7', '#BB5B6A']
gene_df.plot(x="Genes", 
             y=["Present", "Absent"], 
             kind="barh", 
             stacked=True, 
             color=palette[2:]) # Using the last two colors

plt.xlabel('Count')
plt.title('Presence of each resistance gene across the samples')

# Moving the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show();

# This code plots a horizontal stacked bar chart to visualize the presence of each resistance gene
# ('Present' and 'Absent') across the samples in the 'gene_df' DataFrame.
```

## 3.3 Distribution of cell counts for each organism

### using original data

```python
# This code creates a 5x2 grid of subplots and populates each subplot with a histogram overlaid
# with a Kernel Density Estimate (KDE) plot for different columns ('Organism_1' to 'Organism_10')
# of the DataFrame 'df'. The histograms show the distribution of cell counts for each organism.


fig, axes = plt.subplots(5, 2, figsize=(15, 20))  # 5 rows, 2 columns

axes = axes.ravel()  # Flatten the axes

for i in range(1, 11):
    organism_column = f"Organism_{i}"
    # Plotting the histogram with KDE
    sns.histplot(df[organism_column], bins=20, kde=True, ax=axes[i-1], color=palette[i%4])
    axes[i-1].set_title(f'Distribution of cell counts for {organism_column}')

plt.tight_layout()  # Ensures a nice layout
plt.show();
```

### use log transformed data

```python
# This code extends the previous plot by applying a log transformation to the data before plotting.
# The log transformation is used to visualize the data on a log scale, making it easier to interpret
# distributions that have a wide range of values, and a small constant is added to avoid taking the log
# of zero.


import numpy as np

fig, axes = plt.subplots(5, 2, figsize=(15, 20))  # 5 rows, 2 columns

axes = axes.ravel()  # Flatten the axes

for i in range(1, 11):
    organism_column = f"Organism_{i}"
    # Applying log transformation, adding a small constant to avoid log(0)
    data = np.log1p(df[organism_column])
    # Plotting the histogram with KDE
    sns.histplot(data, bins=20, kde=True, ax=axes[i-1], color=palette[i%4])
    axes[i-1].set_title(f'Distribution of cell counts for {organism_column} (log scale)')

plt.tight_layout()  # Ensures a nice layout
plt.show();
```

**Summary**

the distrution of cell counts for each organism is highlyskewed with a long tail to the right. 


# 3.4 Add new data to dataframe

```python
def add_data_to_dataframe(df, data):
    """
    Adds a new row of data to the dataframe.

    Parameters:
    df (pd.DataFrame): The original dataframe.
    data (list): The new data to add. Should have the same number of elements as df has columns.

    Returns:
    pd.DataFrame: The updated dataframe.
    """
    # Check that data has the correct length
    if len(data) != len(df.columns):
        raise ValueError(f"data should have {len(df.columns)} elements")

    # Append the data
    new_df = pd.concat([df, pd.DataFrame([data], columns=df.columns)], ignore_index=True)

    return new_df

```

```python
# Usage example
df = pd.DataFrame({
    'Organism_1': [10, 15, 8],
    'Antibiotic_1': ['R', 'S', 'R'],
    'Gene_1': [0, 1, 1]
})
data = [7, 'S', 0]
new_df = add_data_to_dataframe(df, data)
print(new_df)
```
